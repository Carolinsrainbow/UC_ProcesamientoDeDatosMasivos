1.  Utiliza Python Extraer información, desde la página web:
http://books.toscrape.com/


```python
import requests
from bs4 import BeautifulSoup

url = "http://books.toscrape.com/"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

#print(soup.prettify())
```

2.  Examina el código fuente de la página web:

El método para hacerlo depende del navegador que estés utilizando. Considera que esta página contiene un catálogo de libros, el catálogo cuenta con un total de 1000 libros, pero se muestran 20 a la vez, por lo que para revisarlos todos debes hacer repetidamente click en el botón “next”.

3.  Escribe un programa en python que imprima en la consola:

Los links de todos los libros que tengan una calificación de 5 estrellas en la página 1 de http://books.toscrape.com/


```python
books = soup.find_all('article', class_='product_pod')

five_star_books = []
for book in books:
    if "star-rating Five" in str(book):
        book_link = book.find("h3").find("a")["href"]
        absolute_link = requests.compat.urljoin(url, book_link)
        five_star_books.append(absolute_link)

print("Links de los libros con calificación de 5 estrellas para la página 1:")
for link in five_star_books:
    print(link)
```

4.  Extiende tu programa para extraer e imprimir:

Todos los libros del catálogo completo (las 50 páginas) cuyo rating sea igual a 5 estrellas.

```python
def completeSearch(url):
    results = []

    while url:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")

        books = soup.find_all('article', class_='product_pod')
        for book in books:
            if "star-rating Five" in str(book):
                book_link = book.find("h3").find("a")["href"]
                absolute_link = requests.compat.urljoin(url, book_link)
                results.append(absolute_link)

        next_button = soup.find("li", class_="next")
        if next_button:
            url = requests.compat.urljoin(url, next_button.find("a")["href"])
        else:
            url = None

    return results

complete5star = completeSearch(url)

print("Links de todos los libros con calificación de 5 estrellas:")
for link in complete5star:
    print(link)
```
